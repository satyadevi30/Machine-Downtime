## Cleaning the data

```python
dataset=pd.DataFrame(dataset1['Downtime'].replace(['Machine_Failure', 'No_Machine_Failure'], [1, 0], inplace = True))
```

```python
dataset['Hydraulic_Pressure(bar)'].fillna(dataset['Hydraulic_Pressure(bar)'].mean(), inplace=True)
dataset['Coolant_Pressure(bar)'].fillna(dataset['Coolant_Pressure(bar)'].mean(), inplace=True)
dataset['Air_System_Pressure(bar)'].fillna(dataset['Air_System_Pressure(bar)'].mean(), inplace=True)
dataset['Coolant_Temperature'].fillna(dataset['Coolant_Temperature'].mean(), inplace=True)
dataset['Hydraulic_Oil_Temperature(°C)'].fillna(dataset['Hydraulic_Oil_Temperature(°C)'].mean(), inplace=True)
dataset['Spindle_Bearing_Temperature(°C)'].fillna(dataset['Spindle_Bearing_Temperature(°C)'].mean(), inplace=True)
dataset['Spindle_Vibration(µm)'].fillna(dataset['Spindle_Vibration(µm)'].mean(), inplace=True)
dataset['Tool_Vibration(µm)'].fillna(dataset['Tool_Vibration(µm)'].mean(), inplace=True)
dataset['Spindle_Speed(RPM)'].fillna(dataset['Spindle_Speed(RPM)'].mean(), inplace=True)
dataset['Voltage(volts)'].fillna(dataset['Voltage(volts)'].mean(), inplace=True)
dataset['Torque(Nm)'].fillna(dataset['Torque(Nm)'].mean(), inplace=True)
dataset['Cutting(kN)'].fillna(dataset['Cutting(kN)'].mean(), inplace=True)
```

```python
# Convert 'date' to datetime
dataset1['Date'] = pd.to_datetime(dataset1['Date'])
dataset1['month'] = dataset1['Date'].dt.month
# Filter to get only failure events
failures = dataset[dataset1['Downtime'] == 'Machine_Failure']
# Extract the month from the date
failures['month'] = failures['Date'].dt.month
# Count the number of failures by month
failure_counts = failures['month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
failure_counts.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Number of Failures')
plt.title('Number of Machine Failures by Month')
plt.xticks(range(12), [
    'January', 'February', 'March', 'April', 'May', 
    'June', 'July', 'August', 'September', 'October', 
    'November', 'December'], rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Find the month with the maximum number of failures
max_failures_month = failure_counts.idxmax()
max_failures_count = failure_counts.max()
print(max_failures_month,'-',max_failures_count)
```
